```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Sample dataset: [Years of Experience, Salary]
data = np.array([
    [1, 25000],
    [2, 28000],
    [3, 35000],
    [4, 40000],
    [5, 45000],
    [6, 50000],
    [7, 55000],
    [8, 60000],
    [9, 70000],
    [10, 80000]
])
```

```python
# X should be 2D (n_samples, n_features)
x = data[:, 0].reshape(-1, 1)  # shape becomes (10, 1)
y = data[:, 1]                 # shape is (10,)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predict line
m = model.coef_[0]
c = model.intercept_

# Generate line for plotting
x_line = np.arange(1,11,1)
y_line = m*x_line+c

##   ----If we not use the Predict----------
# x_line = np.arange(1,11).reshape(-1,1)
# y_line = model.predict(x_line)           ##

# Plot the regression line
plt.plot(x_line, y_line, color='red', label='Regression Line')

# Plot training data
plt.scatter(X_train, y_train, color='blue', label='Training Data')

# Plot test data
plt.scatter(X_test, y_test, color='green', label='Test Data')

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression: Salary vs Experience")
plt.legend()
plt.show()
```
