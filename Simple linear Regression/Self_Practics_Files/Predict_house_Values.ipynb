#### Prediction of the House random created data for the visualization of the Data and it outcome

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 1. Sample Dataset
data = {
    'Area_in_sqft': [1000, 1500, 2000, 2500, 3000],
    'Price_in_Lakhs': [50, 75, 100, 130, 150]
}

df = pd.DataFrame(data)

# 2. Features and Target
X = df[['Area_in_sqft']]    # input (2D)
y = df['Price_in_Lakhs']    # target (1D)

# 3. Split data (70% train, 30% test)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Create and train the model
model = LinearRegression()
model.fit(x_train, y_train)

# 5. Make predictions
y_predict = model.predict(x_test)

# 6. Evaluate model (Cost Function + RÂ² Score)
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)

print("Coefficient (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
print("Mean Squared Error (Cost):", round(mse, 2))
print("RÂ² Score:", round(r2, 2))

# 7. Plot
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_predict, color='red', label='Prediction Line')
plt.xlabel('Area (sqft)')
plt.ylabel('Price (Lakh â¹)')
plt.legend()
plt.grid()
plt.title('House Price Prediction')
plt.show()
```
