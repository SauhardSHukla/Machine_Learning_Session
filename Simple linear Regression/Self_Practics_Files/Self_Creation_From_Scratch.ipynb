```python

# 1. Sample Dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# 2. Prediction Function
def predict(x, m, c):
    return m * x + c

# 3. Cost Function (Mean Squared Error)
def cost_function(x, y, m, c):
    y_pred = predict(x, m, c)
    error = y - y_pred
    cost = np.mean(error ** 2)
    return cost

# 4. Gradient Descent - Training Function
def fit(x, y, learning_rate=0.01, iterations=100):
    m = 0
    c = 0
    cost_history = []

    n = len(y)

    for i in range(iterations):
        y_pred = predict(x, m, c)
        error = y - y_pred

        # Gradients (partial derivatives)
        dm = (-2/n) * sum(x * error)
        dc = (-2/n) * sum(error)

        # Update m and c
        m = m - learning_rate * dm
        c = c - learning_rate * dc

        # Save cost at each iteration
        cost = cost_function(x, y, m, c)
        cost_history.append(cost)

        if i % 10 == 0:
            print(f"Iteration {i}: Cost = {round(cost, 4)}")

    return m, c, cost_history

# 5. Train the Model
m, c, cost_history = fit(x, y)

# 6. Final Model Info
print("\nFinal Slope (m):", round(m, 3))
print("Final Intercept (c):", round(c, 3))
print("Final Cost:", round(cost_history[-1], 4))

# 7. Plot Cost vs Iteration
plt.plot(range(len(cost_history)), cost_history, color='green')
plt.title("Cost Over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()

# 8. Prediction and Plot
# x_pred = np.array([6, 7])
# y_pred = predict(x_pred, m, c)
# print("\nPredicted values for", x_pred, "are", y_pred)

# Plot Regression Line
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, predict(x, m, c), color='red', label='Regression Line')
plt.legend()
plt.grid()
plt.title("Linear Regression Fit")
plt.show()
```
