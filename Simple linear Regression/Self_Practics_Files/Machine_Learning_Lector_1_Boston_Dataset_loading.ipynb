```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn import datasets

digits = datasets.load_digits()
type(digits)

x= digits.data
y=digits.target

x_train,x_test ,Y_train,Y_test = model_selection.train_test_split(x,y)

"""Finding the shape of the All the Test and Train"""
# print(x_train,x_test ,Y_train,Y_test)
# print(x_train.shape)
# print(x_test.shape)
# print(Y_train.shape)
# print(Y_test.shape)

# Create and train the model
algo = LinearRegression()
algo.fit(x_train, Y_train)

# Predict values
Y_predict = algo.predict(x_test)

# Scatter plot of actual vs predicted values
plt.scatter(Y_test, Y_predict, color='blue', alpha=0.5)
plt.xlabel("Actual Values (Y_test)")
plt.ylabel("Predicted Values (Y_predict)")
plt.title("Actual vs Predicted Values")

# Fix axis limits
plt.xlim([ -1,10])
plt.ylim([ -1,10])

# Draw a diagonal reference line (ideal predictions)
plt.plot([-1, 10], [-1, 10], color='red', linestyle='dashed')  
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import datasets, model_selection
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# Step 1: Load Dataset
digits = datasets.load_digits()
x = digits.data
y = digits.target

# Step 2: Split Dataset
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, test_size=0.25, random_state=42)

# Step 3: Create and Train the Model
model = LogisticRegression(max_iter=10000)  # Increase max_iter to avoid convergence warnings
model.fit(x_train, y_train)

# Step 4: Predict
y_pred = model.predict(x_test)

# Step 5: Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Step 6: Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
