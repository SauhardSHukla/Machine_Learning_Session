### Code to know the how the working of the code and how to split it 

```python
from sklearn import datasets
from sklearn import  model_selection
from sklearn.model_selection import train_test_split
```

```python
breast_Cancer= datasets.fetch_california_housing()

X = breast_Cancer.data
Y = breast_Cancer.target

import pandas as pd
df = pd.DataFrame(X)
# print(breast_Cancer.feature_names)

df.columns= breast_Cancer.feature_names
df.describe()
```

```python
model_selection.train_test_split(X,Y)
```

```python
x_train,x_test , y_train,y_test = train_test_split(
  X,Y,
  random_state = 42,
  test_size =0.3
)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
algi_1 = LinearRegression()

algi_1.fit(x_train,y_train)

y_predict=algi_1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

print(round(mean_squared_error(y_test,y_predict),3))
print(round(r2_score(y_test,y_predict),3))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5)) 

# Plot predicted values
plt.scatter(range(len(y_predict)), y_predict, color='blue', label='Predicted')

# Plot actual values
plt.scatter(range(len(y_test)), y_test, color='red', label='Actual')

# Add legend, labels, and title
plt.xlabel("Sample Index")
plt.ylabel("Target Value")
plt.title("Predicted vs Actual Values")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot of predicted vs actual
plt.scatter(y_predict, y_test, color='red', label='Predicted vs Actual')

# Add ideal diagonal line (y = x)
min_val = min(min(y_test), min(y_predict))
max_val = max(max(y_test), max(y_predict))
plt.plot([min_val, max_val], [min_val, max_val], color='blue', label='Perfect Prediction Line')

# Labels and legend
plt.xlabel("Predicted Values")
plt.ylabel("Actual Values")
plt.title("Prediction Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
