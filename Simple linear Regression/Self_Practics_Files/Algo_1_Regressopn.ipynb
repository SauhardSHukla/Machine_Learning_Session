```python
# Step 1: Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Step 2: Sample data (Ads vs Sales)
X = np.array([[1], [2], [3], [4], [5],[6],[7],[8]])  # Ads run
y = np.array([10000, 20000, 30000, 40000, 50000,60000,70000,80000])  # Sales 

# Step 3: Train-test split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Create and train model
model = LinearRegression()
model.fit(x_train, y_train)

# Step 5: Predict sales for test data adn r2_score
y_predict = model.predict(x_test)
r2_scor = r2_score(y_test,y_predict)

# Step 6: Show predictions
print("Predicted sales:", y_predict)
print("Actual sales:", y_test)

if r2_scor <= 1.0:
  print('R2_score is best well performed:', r2_scor)

elif r2_scor >= 5.0:
  print('R2_score is good well performed:',r2_scor)
else:
  print('R2_score is bad well performed:',r2_scor)

# Step 7: Plot results
plt.scatter(X, y, color='blue', label='Actual Data')  # original points
plt.plot(X,model.predict(X), color='red', label='Regression Line')  # prediction line

# plt.plot([1,5],[10000,50000], color='red', label='Regression Line')  # Manualy created prediction line

plt.xlabel("No. of Ads Run")
plt.ylabel("Sales ()")
plt.title("Simple Linear Regression: Ads vs Sales")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#m(Change in the sale of the X ) * X(Value of X each time we add new ads) + c(Value of Y when X was '0' ) = Y(Sales) 

print("Co-efficient(Slope = m) ",model.coef_[0])
print('Intercept',model.intercept_)
```
