```python
""" Loading of the Boston Dataset from sklearn for the training"""

from sklearn import datasets
import pandas as pd
```

```python
digits = datasets.load_digits()
type(digits)
```

```python
"""How the Data format looks like"""
# digits
```

```python
x= digits.data
y=digits.target
```

```python
""" Importion of the Pandas for the set the Dataframe """

df=pd.DataFrame(x)
print(digits.feature_names)
df.columns = digits.feature_names
df.describe()
```

```python
from sklearn import model_selection
x_train,x_test ,Y_train,Y_test = model_selection.train_test_split(x,y)
# print(x_train,x_test ,Y_train,Y_test)
# print(x_train.shape)
# print(x_test.shape)
# print(Y_train.shape)
# print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Create and train the model
algo = LinearRegression()
algo.fit(x_train, Y_train)

# Predict values
Y_predict = algo.predict(x_test)

# Scatter plot of actual vs predicted values
plt.scatter(Y_test, Y_predict, color='blue', alpha=0.5)
plt.xlabel("Actual Values (Y_test)")
plt.ylabel("Predicted Values (Y_predict)")
plt.title("Actual vs Predicted Values")

# Fix axis limits
plt.xlim([ -1,10])
plt.ylim([ -1,10])

# Draw a diagonal reference line (ideal predictions)
plt.plot([-1, 10], [-1, 10], color='red', linestyle='dashed')  
plt.show()
```
